```python
import pandas as pd
from glob import glob


stage_order = ['COAD','KIRC','LIHC','LUAD','SKCM','STAD','THCA','UCEC','LUSC','BRCA']
grade_order = ['CESC','KIRC','LGG','LIHC','PAAD','STAD','UCEC','HNSC']
```

```python
csvs = glob('*.csv')
```

```python
stage_res = []
grade_res = []
for csv in csvs:

    info = csv.split('_')
    model = info[0]
    task_type = info[-1].replace('.csv', '')

    dfres = pd.read_csv(csv,index_col=0)
    dfr = dfres.groupby('task').mean().sort_values('score',ascending=False)

    dfr[model] = dfr.score
    dfr.index = dfr.index.map(lambda x:x.split('_')[-2])

    if task_type == 'stage':
        order = stage_order
        dfr = dfr.loc[order]
        stage_res.append(dfr[model])
        
    else:
        order = grade_order
        dfr = dfr.loc[order]
        grade_res.append(dfr[model])    
```

```python
pd.concat(stage_res,axis=1)
```

```python
pd.concat(grade_res,axis=1)
```
