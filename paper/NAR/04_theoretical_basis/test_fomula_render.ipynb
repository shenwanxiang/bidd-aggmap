```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import sys,os
import pandas as pd
import numpy as np
import warnings

sns.set(style = 'white', font_scale=1.5)
```

```python
df1 = pd.read_csv('./history_mnist.csv', index_col='Epochs')
df2 = pd.read_csv('./history_fmnist.csv', index_col='Epochs')
```

```python
dfp1 = df1['Loss'].to_frame(name = 'MNIST').join(df2['Loss'].to_frame(name = 'FMNIST'))
dfp2 = df1['PCC'].to_frame(name = 'MNIST').join(df2['PCC'].to_frame(name = 'FMNIST'))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 6))
dfp1.rolling(1).mean().plot(ax = ax1, color= ['blue', 'red'])
dfp2.rolling(1).mean().plot(ax = ax2, color= ['blue', 'red'])

ax1.set_ylim(10.35, 10.8)
ax2.set_ylim(0.45, 0.575)

ax1.set_ylabel('CE_Loss')
ax2.set_ylabel('PCC')

ax1.set_xlim(0, 500)
ax2.set_xlim(0, 500)

ax1.tick_params(axis='y',which='major', direction='out', left=True, length=8, color='black') 
ax1.tick_params(axis='x', which='major', direction='out',bottom=True, length=8, color='black') 

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)


ax2.tick_params(axis='y',which='major', direction='out', left=True, length=8, color='black') 
ax2.tick_params(axis='x', which='major', direction='out',bottom=True, length=8, color='black') 

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

fig.tight_layout()

fig.savefig('./mninst_fmnist.pdf', dpi = 400)
```
