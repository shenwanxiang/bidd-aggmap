```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from aggmap.utils.matrixopt import conv2
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
```

```python
model = 'MP' #, model = 'Stage_0'
res = []
for i in range(9):
    df = pd.read_excel('./%s_model/feature_imp_8_%s.xlsx' % ( model,i))
    df = df[['x', 'y', 'v', 'Subtypes', 'colors', 'MP_importance']]
    s = np.log(df[['%s_importance' % model]])
    a = scaler.fit_transform(s.values)
    a = a.reshape(86, 85)
    IMPM = conv2(a, kernel_size=5, sigma=1)
    df['%s_importance_fold_%s' % (model, i)] = IMPM.reshape(-1,)
    df.to_excel('./%s_model/feature_imp_8_%s.xlsx' % ( model,i))

    sns.set(style='white',  font='sans-serif', font_scale=2)
    fig, ax  = plt.subplots(1, 1, figsize=(10, 9))

    sns.heatmap(IMPM,  
                cmap = 'rainbow', alpha = 0.8, xticklabels=10, ax =ax,
                yticklabels=10, vmin = 0.2,
                cbar_kws = {'fraction':0.046, 'pad':0.02})
    res.append(df['%s_importance_fold_%s' % (model, i)])
    
    
IMPM = pd.concat(res,axis=1).mean(axis=1).values.reshape(86, 85)
sns.set(style='white',  font='sans-serif', font_scale=2)
fig, ax  = plt.subplots(1, 1, figsize=(10, 9))

sns.heatmap(IMPM,  
            cmap = 'rainbow', alpha = 0.8, xticklabels=10, ax =ax,
            yticklabels=10, vmin = 0.2,
            cbar_kws = {'fraction':0.046, 'pad':0.02})
```
